```python
commercial_comments_df = pd.read_excel('s3://elektra-data/commercial_comments/20240722 Commercial Comments_ Price Recomendations July 22nd 2024.xlsx',header=2)
commercial_comments_df

filtered = commercial_comments_df[['sku_number','Estatus','Precio DE ($)','Precio A ($)','Precio A ($).1','Costo Unitario','Margen Objetivo Neto Telefon√≠a']]

filtered['Control Cash Precio A'] = filtered['Precio A ($)']
filtered['Control Credit Precio A'] = filtered['Precio A ($).1']
filtered['Control Precio DE ($)'] = filtered['Precio DE ($)']
filtered['Cost of Sale Credit'] = filtered['Costo Unitario']
filtered['Margin'] = filtered['Margen Objetivo Neto Telefon√≠a']
filtered['SKU'] = filtered['sku_number']
filtered.drop(['Precio A ($)','Precio A ($).1','Precio DE ($)','Margen Objetivo Neto Telefon√≠a','Costo Unitario','sku_number'],axis=1,inplace=True)

def assign_price_segment(df, num_segments):
    
    # Check if the necessary columns exist
    if 'SKU' not in df.columns or 'Control Credit Precio A' not in df.columns:
        raise ValueError("The CSV file must contain 'SKU' and 'Price' columns.")
    
    # Assign price segments using pd.qcut
    df['price_segment'] = pd.qcut(df['Control Credit Precio A'], num_segments, labels=False) + 1
    
    return df

df_with_segments = assign_price_segment(filtered, num_segments=10)
# df_with_segments.sort_values('Control Credit Precio A').tail(50)
# df_with_segments.to_csv('data/national_price_20240726.csv')
```

```python
import boto3
import pandas as pd
from io import StringIO

def save_csv_to_s3(df, bucket_name, s3_file_path):
    # Convert DataFrame to CSV
    csv_buffer = StringIO()
    df.to_csv(csv_buffer, index=False)

    # Get the contents of the buffer
    csv_buffer.seek(0)
    
    # Create a session using boto3
    s3 = boto3.client('s3')

    # Upload the CSV file to S3
    s3.put_object(Bucket=bucket_name, Key=s3_file_path, Body=csv_buffer.getvalue())
    print(f"CSV file uploaded successfully to s3://{bucket_name}/{s3_file_path}")

bucket_name = 'elektra-data'
s3_file_path = 'commercial_comments/national_price_20240722.csv'
save_csv_to_s3(df_with_segments, bucket_name, s3_file_path)
```

# Validation Check

```python
import pandas as pd
def validation_check_price_recommendations(commercial_comments_path: str, predictions_path: str):
    # Read the Excel file using Pandas
    commercial_comments_df = pd.read_excel(commercial_comments_path, header=4)

    # Select and rename columns
    recommendations_df = commercial_comments_df[['sku_number', 'Estatus', 'Precio DE ($)', 'Precio A ($)', 'Precio A ($).1', 'Costo Unitario', 'Margen Objetivo Neto Telefon√≠a']].rename(columns={
        'Precio A ($)': 'Control Cash Precio A',
        'Precio A ($).1': 'Control Credit Precio A',
        'Precio DE ($)': 'Control Precio DE ($)',
        'Costo Unitario': 'Cost of Sale Credit',
        'Margen Objetivo Neto Telefon√≠a': 'Margin',
        'sku_number': 'SKU'
    })

    # Read the CSV file using Pandas
    final_prices = pd.read_csv(predictions_path)

    # Merge the DataFrames
    final_prices = final_prices.merge(recommendations_df[['SKU', 'Control Cash Precio A', 'Control Precio DE ($)', 'Cost of Sale Credit']], on='SKU', how='inner')

    # Find inequalities and print results
    inequal_A = final_prices[final_prices['Best Price Credit A'] != final_prices['Best Price Cash A']]
    print("Inequal Price A in Final Spreadsheet: " + str(len(inequal_A)))

    inequal_DE = final_prices[final_prices['Best Price Credit DE'] != final_prices['Best Price Cash DE']]
    print("Inequal Price DE in Final Spreadsheet: " + str(len(inequal_DE)))

    inequal_DE_national = final_prices[final_prices['Best Price Credit DE'] != final_prices['Control Precio DE ($)']]
    print("Inequal Price DE between national and US in Final Spreadsheet: " + str(len(inequal_DE_national)))

    de_less_A = final_prices[(final_prices['Best Price Credit DE'] <= final_prices['Best Price Credit A']) | (final_prices['Best Price Cash DE'] <= final_prices['Best Price Cash A'])]
    print("DE < A in Final Spreadsheet: " + str(len(de_less_A)))

    dates = final_prices['Date'].nunique()
    print("Number of distinct dates in final prices: " + str(dates))

    null_values = final_prices.isnull().sum().sum()
    print("Null values: " + str(null_values))

commercial_comments_path = 's3://elektra-data/commercial_comments/Commercial Comments Price Recommendations July 25 2024.xlsx'
predictions_path = 'data/predictions/07302024_EKT_Physical_Stores_Prices_Final.csv'

validation_check_price_recommendations(commercial_comments_path, predictions_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('Analysis_July30.csv')  # Replace with your actual file path
df['Cost of Sale Credit'] = df['Cost of Sale Credit_x']
# Function to calculate percentage difference
def percent_diff(a, b):
    return (a - b) / b * 100

# 1. Compare 'Old Best Price A' and 'Best Price Credit A'
df['price_change_percent'] = percent_diff(df['Best Price Credit A'], df['Old Best Price A'])

# 2. Analyze price changes by price segment, brand, and Estatus
segments = ['price_segment', 'Brand', 'Estatus']
for segment in segments:
    print(f"\nAverage price change by {segment}:")
    print(df.groupby(segment)['price_change_percent'].mean().sort_values(ascending=False))

# 3. Margin analysis
df['margin'] = df['Best Price Credit A'] - df['Cost of Sale Credit']
df['margin_percent'] = df['margin'] / df['Best Price Credit A'] * 100

print("\nAverage margin percentage by Brand:")
print(df.groupby('Brand')['margin_percent'].mean().sort_values(ascending=False))

# 4. Compare 'Best Price Credit A' and 'Control Cash Precio A'
df['pilot_control_diff_percent'] = percent_diff(df['Best Price Credit A'], df['Control Cash Precio A'])

print("\nAverage difference between 'Best Price Credit A' and 'Control Cash Precio A':")
print(df['pilot_control_diff_percent'].mean())

print("\nAverage 'Best Price Credit A' vs 'Control Cash Precio A' difference by Brand:")
print(df.groupby('Brand')['pilot_control_diff_percent'].mean().sort_values(ascending=False))

# 5. Visualizations

# Price change distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['price_change_percent'], kde=True)
plt.title('Distribution of Price Changes (Old vs New Credit Price)')
plt.xlabel('Percentage Change')
plt.savefig('price_change_distribution.png')
plt.close()

# Margin by Brand
plt.figure(figsize=(12, 6))
sns.boxplot(x='Brand', y='margin_percent', data=df)
plt.title('Margin Percentage by Brand')
plt.xticks(rotation=45)
plt.savefig('margin_by_brand.png')
plt.close()

# Credit A vs Cash A difference by Brand
plt.figure(figsize=(12, 6))
sns.boxplot(x='Brand', y='pilot_control_diff_percent', data=df)
plt.title('Pilot  vs Control Price Difference by Brand')
plt.xticks(rotation=45)
plt.savefig('pilot_control_diff_by_brand.png')
plt.close()

# 6. Summary statistics
summary = df.groupby('Brand').agg({
    'price_change_percent': 'mean',
    'margin_percent': 'mean',
    'pilot_control_diff_percent': 'mean'
}).round(2)

summary.columns = ['Avg Price Change %', 'Avg Margin %', 'Avg Control VS Pilot Diff %']
summary = summary.sort_values('Avg Margin %', ascending=False)

print("\nSummary Statistics by Brand:")
print(summary)

# Save summary to CSV
summary.to_csv('brand_summary.csv')

# 7. Additional analysis on 'Best Price Credit A' vs 'Control Cash Precio A'
print("\nPercentage of products where 'Best Price Credit A' is more expensive:")
print((df['pilot_control_diff_percent'] > 0).mean() * 100)

print("\nAverage percentage difference when 'Best Price Credit A' is more expensive:")
print(df[df['pilot_control_diff_percent'] > 0]['pilot_control_diff_percent'].mean())

print("\nAverage percentage difference when 'Best Price Credit A' is cheaper:")
print(df[df['pilot_control_diff_percent'] < 0]['pilot_control_diff_percent'].mean())

# Visualize the distribution of differences
plt.figure(figsize=(10, 6))
sns.histplot(df['pilot_control_diff_percent'], kde=True)
plt.title('Distribution of Price Differences (Pilot A vs Contrl A)')
plt.xlabel('Percentage Difference')
plt.axvline(x=0, color='r', linestyle='--')
plt.savefig('pilot_control_diff_distribution.png')
plt.close()

print("\nAnalysis complete. Check the generated CSV file and PNG images for visualizations.")
```
